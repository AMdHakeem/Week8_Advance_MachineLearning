```python
# Importing Required Libraries for Data Analysis and Machine Learning

import pandas as pd
from sklearn.model_selection import train_test_split 
import time
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import pickle
import matplotlib.pyplot as plt


# Feature Selection Using SelectKBest Method
def selectkbest(indep_X,dep_Y,n):
        test = SelectKBest(score_func=chi2, k=n)
        fit1= test.fit(indep_X,dep_Y)
        selectk_features = fit1.transform(indep_X)
        return selectk_features

# Explanation: This function selects the top n important features from the dataset using SelectKBest with the Chi-Square Test. 
# It checks which input features have the strongest relationship with the target variable and keeps only those features for further model training.
    
def split_scalar(indep_X,dep_Y):
        X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size = 0.25, random_state = 0)
        sc = StandardScaler()
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)    
        return X_train, X_test, y_train, y_test

# Explanation: This function splits the dataset into training and testing sets using scikit-learn train_test_split, where 75% is used for training and 25% for testing the model.
# It then applies StandardScaler to normalize the feature values so that all variables are on the same scale.
# Finally, the scaled training and testing data are returned for use in machine learning models.
 
# Model Prediction Function
def cm_prediction(classifier,X_test):
     y_pred = classifier.predict(X_test)

# Explanation: This function uses the trained classifier to predict the output values for the test dataset.
# It applies the model’s predict() method to generate predicted labels based on the input test features.
# The predicted results are stored in y_pred for further evaluation such as accuracy or confusion matrix.
        
# Making the Confusion Matrix
     from sklearn.metrics import confusion_matrix
     cm = confusion_matrix(y_test, y_pred)

# Explanation: This code imports the Confusion Matrix from scikit-learn.
# It compares the actual values (y_test) with the predicted values (y_pred) to evaluate the model’s classification performance.
# The confusion matrix shows the number of correct and incorrect predictions made by the model.

# Importing Model Evaluation Metrics
     from sklearn.metrics import accuracy_score 
     from sklearn.metrics import classification_report 

# Explanation: This code imports Accuracy Score and Classification Report from scikit-learn.
# These metrics are used to evaluate the performance of a classification model.
# Accuracy measures the percentage of correct predictions, 
# while the classification report shows precision, recall, and F1-score for each class.

# Model Accuracy and Performance Report
     Accuracy=accuracy_score(y_test, y_pred )

# This code calculates the model’s performance using Accuracy Score by comparing the actual values (y_test) with the predicted values (y_pred).
# It also generates a Classification Report, which shows precision, recall, and F1-score for each class.
# Finally, it returns the classifier, accuracy value, report, test data, and confusion matrix for further analysis.

# Generating Classification Report and Returning Results
     report=classification_report(y_test, y_pred)
     return  classifier,Accuracy,report,X_test,y_test,cm

# Explanation: This code creates a Classification Report by comparing the actual test values (y_test) with the predicted values (y_pred).
# The report provides important metrics like precision, recall, and F1-score to evaluate the model’s performance.
# Finally, it returns the classifier, accuracy, report, test data, and confusion matrix for further analysis.

# Logistic Regression Model Training
def logistic(X_train,y_train,X_test):       
        # Fitting K-NN to the Training set
        from sklearn.linear_model import LogisticRegression
        classifier = LogisticRegression(random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm   
    
# This function trains a classification model using Logistic Regression from scikit-learn.
# The model is fitted using the training data (X_train, y_train) to learn patterns in the dataset.
# After training, it calls the prediction function to calculate accuracy, generate the report, and return the evaluation results.

# Support Vector Machine (Linear Kernel) Model Training
def svm_linear(X_train,y_train,X_test):                
        from sklearn.svm import SVC
        classifier = SVC(kernel = 'linear', random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm
    
# This function trains a classification model using Support Vector Machine with a linear kernel from scikit-learn.
# The model learns patterns from the training data (X_train, y_train) to classify the input features.
# After training, it predicts the test data and returns the accuracy, classification report, and confusion matrix for evaluation.

# Support Vector Machine (Non-Linear Kernel) Model Training    
def svm_NL(X_train,y_train,X_test):
        from sklearn.svm import SVC
        classifier = SVC(kernel = 'rbf', random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm

# This function trains a model using Support Vector Machine with a non-linear RBF kernel from scikit-learn.
# The model learns complex patterns from the training data (X_train, y_train) to improve classification performance.
# After training, it predicts the test data and returns the accuracy, classification report, and confusion matrix for evaluation.

# Naive Bayes Model Training
def Navie(X_train,y_train,X_test):       
        # Fitting K-NN to the Training set
        from sklearn.naive_bayes import GaussianNB
        classifier = GaussianNB()
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm         

# This function trains a classification model using Naive Bayes with the GaussianNB method from scikit-learn.
# The model learns patterns from the training data (X_train, y_train) based on probability distribution.
# After training, it predicts the test data and returns the accuracy, classification report, and confusion matrix for evaluation.

# K-Nearest Neighbors (KNN) Model Training
def knn(X_train,y_train,X_test):           
        # Fitting K-NN to the Training set
        from sklearn.neighbors import KNeighborsClassifier
        classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm

# This function trains a classification model using the K-Nearest Neighbors from scikit-learn.
# The model classifies new data by comparing it with the 5 nearest neighbors based on distance.
# After training, it predicts the test data and returns the accuracy, classification report, and confusion matrix for evaluation.

# Decision Tree Model Training
def Decision(X_train,y_train,X_test):        
        # Fitting K-NN to the Training set
        from sklearn.tree import DecisionTreeClassifier
        classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm      

# This function trains a classification model using the Decision Tree from scikit-learn.
# The model learns decision rules from the training data (X_train, y_train) using the entropy criterion to split the data.
# After training, it predicts the test data and returns the accuracy, classification report, and confusion matrix for evaluation.    

# Random Forest Model Training
def random(X_train,y_train,X_test):        
        # Fitting K-NN to the Training set
        from sklearn.ensemble import RandomForestClassifier
        classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm

# This function trains a classification model using the Random Forest from scikit-learn.
# The model builds multiple decision trees and combines their results to improve prediction accuracy.
# After training, it predicts the test data and returns the accuracy, classification report, and confusion matrix for evaluation.
    
# Model Accuracy Comparison Table
def selectk_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf):     
    dataframe=pd.DataFrame(index=['ChiSquare'],columns=['Logistic','SVMl','SVMnl','KNN','Navie','Decision','Random'])
    for number,idex in enumerate(dataframe.index):      
        dataframe.loc[idex, 'Logistic'] = acclog[number]
        dataframe.loc[idex, 'SVMl'] = accsvml[number]
        dataframe.loc[idex, 'SVMnl'] = accsvmnl[number]
        dataframe.loc[idex, 'KNN'] = accknn[number]
        dataframe.loc[idex, 'Navie'] = accnav[number]
        dataframe.loc[idex, 'Decision'] = accdes[number]
        dataframe.loc[idex, 'Random'] = accrf[number]
    return dataframe

# This function creates a pandas DataFrame to compare the accuracy of different machine learning models.
# It stores the accuracy values of models like Logistic Regression, SVM, KNN, Naive Bayes, Decision Tree, and Random Forest.
# Finally, it returns a table that shows the performance of each model for easy comparison.
```

```python
# Dataset Loading and Data Preprocessing
dataset1=pd.read_csv("prep.csv",index_col=None)

df2=dataset1

df2 = pd.get_dummies(df2, drop_first=True)

indep_X=df2.drop('classification_yes', axis=1)
dep_Y=df2['classification_yes']

# This code loads the dataset from a CSV file using pandas and stores it in a DataFrame.
# It converts categorical variables into numerical values using get_dummies() so they can be used in machine learning models.
# Finally, it separates the data into independent features (indep_X) and the target variable (dep_Y) for model training.
```

```python
# Feature Selection and Accuracy List Initialization
kbest=selectkbest(indep_X,dep_Y,6)       

acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accnav=[]
accdes=[]
accrf=[]

# This code selects the top 6 important features from the dataset using the previously defined SelectKBest function.
# It then creates empty lists to store the accuracy results of different machine learning models.
# These lists will later be used to compare the performance of all models.
```

```python
# Displaying Selected Features
kbest
```

```python
# Model Training and Accuracy Comparison

X_train, X_test, y_train, y_test=split_scalar(kbest,dep_Y)     
        
classifier,Accuracy,report,X_test,y_test,cm=logistic(X_train,y_train,X_test)
acclog.append(Accuracy)

classifier,Accuracy,report,X_test,y_test,cm=svm_linear(X_train,y_train,X_test)  
accsvml.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=svm_NL(X_train,y_train,X_test)  
accsvmnl.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=knn(X_train,y_train,X_test)  
accknn.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=Navie(X_train,y_train,X_test)  
accnav.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=Decision(X_train,y_train,X_test)  
accdes.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=random(X_train,y_train,X_test)  
accrf.append(Accuracy)
    
result=selectk_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf)

# This code splits the selected features into training and testing data, then trains multiple machine learning models using scikit-learn.
# Models such as Logistic Regression, SVM, KNN, Naive Bayes, Decision Tree, and Random Forest are applied to the dataset.
# The accuracy of each model is stored and finally combined into a table to compare their performance.
```
